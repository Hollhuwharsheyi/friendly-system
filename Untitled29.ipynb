```python
# List to store the scraped data
booktitles=[]
booklinks=[]
Authors =[]
year1s=[]

for j in range(0,13):
    url = ('https://pureportal.coventry.ac.uk/en/organisations/school-of-economics-finance-and-accounting/publications/?page={0}'.format(j))
    page = requests.get(url)
    s = BeautifulSoup(page.text, "html.parser") 
    
    all_staff = s.find_all("div", class_="result-container")

    for first_staff in all_staff:
        try:
            booktitle = first_staff.find("h3", class_="title").text
            booktitles.append(booktitle)
        except Exception:
                    booktitle = 'None' 
                    booktitles.append(Author)
                    
        try:
            booklink = first_staff.find( href=True)["href"]
            booklinks.append(booklink)
        except Exception:
                    booklink = 'None' 
                    booklinks.append(Author)
                    
        try:
            year1=first_staff.find('span', class_='date').text
            year1s.append(year1)
        except Exception:
                    year1 = 'None' 
                    year1s.append(Author)
    
        try:
             Author = first_staff.find('a', class_='link person').text 
             Authors.append(Author)
        except Exception:
                    Author = 'None' 
                    Authors.append(Author)
            
       
    
    print(booktitles, booklinks, Authors,year1s)      
        
    

    
```

```python
import pandas as pd
test_df = pd.DataFrame({'booktitles': booktitles,
'booklinks': booklinks,
'Authors': Authors,
'year1s': year1s
})
print(test_df.info())
test_df.count()
```
